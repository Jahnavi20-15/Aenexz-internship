```python
# 1. Function to calculate the area of a circle
def circle_area(radius):
    return 3.14 * radius * radius

print("Area of circle:", circle_area(5))
```

```python
# 2. Function to find the largest of three numbers
def largest_number(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= c:
        return b
    else:
        return c

print("Largest number:", largest_number(10, 25, 15))
```

```python

# 3. Function to find the length of a string without using len()
def string_length(text):
    count = 0
    for char in text:
        count += 1
    return count

print("Length of string:", string_length("Python"))
```

```python
# 4. Function to check whether a number is prime
def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False

    return True

print("Is 17 prime?", is_prime(17))
```

```python
# 5. Recursive function to find nth Fibonacci number
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print("Fibonacci of 6:", fibonacci(6))
```

```python

# 6. Recursive function to reverse a string
def reverse_string(text):
    if text == "":
        return text
    return reverse_string(text[1:]) + text[0]

print("Reversed string:", reverse_string("Python"))
```

```python
# 7. Program to find the sum of numbers from 1 to n
n = 10
total = 0

for i in range(1, n + 1):
    total += i

print("Sum from 1 to", n, "=", total)
```

```python

# 8. Program to count even numbers in a list
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
even_count = 0

for num in numbers:
    if num % 2 == 0:
        even_count += 1

print("Count of even numbers:", even_count)
```

```python
# 9. Program to print prime numbers between 1 and 100
print("Prime numbers between 1 and 100:")

for num in range(2, 101):
    prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            prime = False
            break

    if prime:
        print(num, end=" ")
```
